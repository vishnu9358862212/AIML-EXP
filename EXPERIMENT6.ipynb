```python
import pandas as pd
from google.colab import files
uploaded=files.upload()
```

```python
import io
df=pd.read_csv(io.BytesIO(uploaded['organizations-100.csv']))
```

Remove rows or columns with missing data (if the amount is small)

```python
# Remove rows with missing values
df.dropna(axis=0, inplace=True)

# Remove columns with missing values
df.dropna(axis=1, inplace=True)
```

Replace missing values with statistical measures

```python
# Fill missing numerical data with mean
df['Number of employees'].fillna(df['Number of employees'].mean(), inplace=True)

# Fill missing numerical data with median
df['Number of employees'].fillna(df['Number of employees'].median(), inplace=True)

# Fill missing categorical data with mode
df['Number of employees'].fillna(df['Number of employees'].mode()[0], inplace=True)
```

For time-series or continuous data, interpolation is a good option

```python
df.infer_objects(copy=False)
```

Carry forward or backward values in time-series data

```python
# Forward fill
df.ffill(inplace=True)

# Backward fill
df.bfill(inplace=True)
```

This method calculates how many standard deviations a point is from the mean

```python
from scipy import stats
import numpy as np

# Z-score
z_scores = np.abs(stats.zscore(df['Number of employees']))
df = df[(z_scores < 3)]  # Keep rows with Z-score less than 3
```

Removing Outliers

```python

Q1 = df['Number of employees'].quantile(0.25)
Q3 = df['Number of employees'].quantile(0.75)
IQR = Q3 - Q1

# Filter out outliers
df = df[~((df['Number of employees'] < (Q1 - 1.5 * IQR)) | (df['Number of employees'] > (Q3 + 1.5 * IQR)))]
```

Limit extreme values to reduce the effect of outliers without removing them

```python
from scipy.stats.mstats import winsorize

df['Number of employees'] = winsorize(df['Number of employees'], limits=[0.05, 0.05])
```

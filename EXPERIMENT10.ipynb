```python
import pandas as pd

# Load the Iris dataset
from sklearn.datasets import load_iris

iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species'] = iris.target
data['species'] = data['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Display the dataset's structure
print("Dataset Structure:")
print(data.info())
print("\nFirst Few Rows:")
print(data.head())
```

```python
import matplotlib.pyplot as plt
```

```python
data.iloc[:, :-1].hist(figsize=(10, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

```python
data['species'].value_counts().plot(kind='bar', color=['red', 'green', 'blue'])
plt.title('Distribution of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
data['species'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['red', 'green', 'blue'])
plt.title('Species Proportion')
plt.ylabel('')  # Removes the y-axis label for a cleaner look
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data.iloc[:, :-1], palette='pastel')
plt.title('Box Plot of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal length (cm)', data=data, palette='muted')
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

```python
sns.pairplot(data, hue='species', palette='husl', diag_kind='kde')
plt.suptitle('Pair Plot of Iris Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr_matrix = data.iloc[:, :-1].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
